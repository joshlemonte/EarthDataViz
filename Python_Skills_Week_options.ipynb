```python
# Option A: Color points by location_group
groups = df["location_group"].astype(str)

plt.figure()
for g in sorted(groups.unique()):
    m = groups == g
    plt.scatter(df.loc[m, "dist_coast_km"], df.loc[m, "cl_mgL"], label=g, alpha=0.8)

plt.yscale("log")
plt.xlabel("Distance to coast (km)")
plt.ylabel("Cl (mg/L) [log scale]")
plt.title("Cl vs distance to coast (colored by location_group)")
plt.legend(fontsize=8, loc="best")
plt.tight_layout()
plt.show()
```

```python
# Option B: Swap y-variable (Na or EC)
yvar = "na_mgL"   # try: "na_mgL" or "ec_uScm"
ylabel = "Na (mg/L)" if yvar == "na_mgL" else "EC (uS/cm)"

plt.figure()
plt.scatter(df["dist_coast_km"], df[yvar])
plt.yscale("log") if yvar != "ec_uScm" else None  # EC may or may not need log; try both
plt.xlabel("Distance to coast (km)")
plt.ylabel(ylabel + (" [log scale]" if yvar != "ec_uScm" else ""))
plt.title(f"{ylabel} vs distance to coast")
plt.tight_layout()
plt.show()
```

```python
# Option C: Filter missing values, then plot again
dff = df.dropna(subset=["dist_coast_km","cl_mgL"]).copy()
print("Rows before:", df.shape[0], "after filtering:", dff.shape[0])

plt.figure()
plt.scatter(dff["dist_coast_km"], dff["cl_mgL"])
plt.yscale("log")
plt.xlabel("Distance to coast (km)")
plt.ylabel("Cl (mg/L) [log scale]")
plt.title("Filtered (no missing distance or Cl)")
plt.tight_layout()
plt.show()
```
